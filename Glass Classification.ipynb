## Data Preprocessing

```python
X = df.drop(['Type', 'Type Name'],axis=1)
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Classification Models

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 2, max_depth = 10, max_features = 5, random_state = 42)
dtc.fit(X_train, y_train)

dtc_pred = dtc.predict(X_test)
```

```python
score = round((accuracy_score(y_test, dtc_pred)*100),2)
print("Algorithm Performed: Decision Tree Classifier")
print("\nModel Score: ", score, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, dtc_pred))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 200, criterion = 'entropy', max_depth = 10, n_jobs = -1, max_features = 3, random_state = 42)
rfc.fit(X_train, y_train)

rfc_pred = rfc.predict(X_test)
```

```python

score = round((accuracy_score(y_test, rfc_pred)*100),2)
print("Algorithm Performed: Random Forest Classifier")
print("\nModel Score: ", score, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, rfc_pred))
    
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knnc = KNeighborsClassifier(n_neighbors = 6, p =1)
knnc.fit(X_train, y_train)

knnc_pred = knnc.predict(X_test)
```

```python
score = round((accuracy_score(y_test, knnc_pred)*100),2)
print("Algorithm Performed: K Nearest Neighbors Classifier")
print("\nModel Score: ", score, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, knnc_pred))
```

### Support Vector Machine

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'linear', random_state = 42)
svc.fit(X_train, y_train)

svc_pred = svc.predict(X_test)
```

```python
score = round((accuracy_score(y_test, svc_pred)*100),2)
print("Algorithm Performed: Support Vector Machine")
print("\nModel Score: ", score, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, svc_pred))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
```

```python
score = round((accuracy_score(y_test, lr_pred)*100),2)
print("Algorithm Performed: Logistic Regression")
print("\nModel Score: ", score, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, lr_pred))
```

## Results

```python
results = pd.DataFrame([
    {'Algorithm' : 'Decision Tree Classifier', 'Accuracy Score (%)' : round((accuracy_score(y_test, dtc_pred)*100),2)},
    {'Algorithm' : 'Random Forest Classifier', 'Accuracy Score (%)' :round((accuracy_score(y_test, rfc_pred)*100),2)},
    {'Algorithm' : 'K Nearest Neighbor', 'Accuracy Score (%)' :round((accuracy_score(y_test, knnc_pred)*100),2)},
    {'Algorithm' : 'Support Vector Machine', 'Accuracy Score (%)' : round((accuracy_score(y_test, svc_pred)*100),2)},
    {'Algorithm' : 'Logistic Regression', 'Accuracy Score (%)' :round((accuracy_score(y_test, lr_pred)*100),2)}
])

results.sort_values(by=['Accuracy Score (%)'], ascending=False)
```
